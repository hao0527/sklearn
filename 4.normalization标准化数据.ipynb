```python
from sklearn import preprocessing
import numpy as np
from sklearn.model_selection import train_test_split    # 分开训练数据和测试数据
from sklearn.datasets.samples_generator import make_classification
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
a = np.array([[10,2.7,3.6],
              [-100,5,-2],
              [120,20,40]],dtype='float64')
print(preprocessing.scale(a))    # 归一化预处理
```

```python
X, y = make_classification(n_samples=300, n_features=2,
                           n_redundant=0, n_informative=2,
                           random_state=22, n_clusters_per_class=1, scale=100)
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
Nor_X = preprocessing.scale(X)    # 归一化以后
Nor_X_train, Nor_X_test, y_train, y_test = train_test_split(Nor_X, y, test_size=.3)
clf = SVC(gamma='auto')
clf.fit(Nor_X_train, y_train)
print(clf.score(Nor_X_test, y_test))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)    # 没有归一化
clf = SVC(gamma='auto')
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```
